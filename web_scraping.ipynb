# All content :

```python
page = 1
while( page<3 ):
    try :
        webpage = requests.get(f'https://quotes.toscrape.com/page/{page}/').text  
    except :
        break
    html = bs.BeautifulSoup(webpage , 'lxml')
    div = html.find_all( 'div' , class_='quote')
    
    for i in div :
        author =  i.find('small' , class_='author')
        quote = i.find( 'span' , class_='text')
        print( author.text , quote.text , sep='\n' )
        tag_div = i.find( 'div' , class_='tags')
        tags = tag_div.find_all( 'a', class_='tag')
        for j in tags :
            print( j.text ,end=' , ')
        print('\n')
    page+=1
```

# For all the quotes on the site :

```python
page=1
while( True ) :
    try :
        webpage = requests.get(f'https://quotes.toscrape.com/page/{page}/').text 
    except :
        break
        
    html = bs.BeautifulSoup(webpage.text , 'lxml')
    if( len(html)==0 ) :
        break
        
    quote = html.find_all( 'div' , class_='quote')
    print( f'For quotes present on Page {page}:\n')
    for i in quote :
        quote = i.find( 'span' , class_='text').text
        author = i.find( 'small' , class_='author').text
        print( quote, author, '\n' , sep=" ---- ")
    page+=1
```

# For number of quotes on the site by an author :

```python
author = 'Albert Einstein'
ans = 0;
page = 1
while( page<11 ):
    try :
        webpage = requests.get(f'https://quotes.toscrape.com/page/{page}/').text  
    except :
        break
    html = bs.BeautifulSoup(webpage , 'lxml')
    quote = html.find_all( 'div' , class_='quote')
    for i in quote :
        if( author == ( i.find( 'small' , class_='author').text ) ):
            ans+=1
    page+=1
print( f'Total number of Quotes found till page 10 are {ans}')
```

# For all the quotes on the site by an author :

```python
author = 'Albert Einstein'
page = 1
while( page<11 ):
    try :
        webpage = requests.get(f'https://quotes.toscrape.com/page/{page}/').text 
    except :
        break
    html = bs.BeautifulSoup(webpage , 'lxml')
    quote = html.find_all( 'div' , class_='quote')
    for i in quote :
        author_found = i.find( 'small' , class_='author').text
        if( author==author_found ):
            quote = i.find( 'span' , class_='text').text
            print( quote,  '\n' )
    page+=1
```

# To get Author & Quote wrt a particular Tag

```python
Tag = 'inspirational'
page = 1
while( page<2 ):
    try :
        webpage = requests.get(f'https://quotes.toscrape.com/page/{page}/').text  
    except :
        break
    html = bs.BeautifulSoup(webpage , 'lxml')
    div = html.find_all( 'div' , class_='quote')
    
    for i in div :
        tag_div = i.find( 'div' , class_='tags')
        tags = tag_div.find_all( 'a', class_='tag')
        for j in tags :
            if( j.text ==Tag ):
                author =  i.find('small' , class_='author')
                quote = i.find( 'span' , class_='text')
                print( author.text , quote.text ,'\n', sep='\n'  )
    page+=1
```

# Top 3 tags present on that page 

```python
Tag = {}
page = 1
try:
    webpage = requests.get(f'https://quotes.toscrape.com/page/{page}/').text  
    html = bs.BeautifulSoup(webpage, 'lxml')
    div = html.find_all('div', class_='quote')
    
    for i in div:
        tag_div = i.find('div', class_='tags')
        tags = tag_div.find_all('a', class_='tag')
        for j in tags:
            if j.text in Tag:
                Tag[j.text] += 1
            else:
                Tag[j.text] = 1

    top_3 = []
    for i in Tag :
        top_3.append([Tag[i],i])
    top_3.sort(reverse=True)
    
    print(f"Top 3 Tags on Page {page}:\n")
    for i in range(3):
        print(f"{top_3[i][1]}  :  {top_3[i][0]} times")

except:
    print("This Page is Not Available")
```
